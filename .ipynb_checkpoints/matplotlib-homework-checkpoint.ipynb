```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns


# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
total_ride_data = city_data.merge(ride_data, how ="left", on=["city", "city"])

# Display the data table for preview
total_ride_data.head()
```

```python
#group by city
ride_summary = total_ride_data.groupby(["city"])
ride_summary.head(1)
```

```python
rides_df = pd.DataFrame({"Driver Count": ride_summary["driver_count"].mean()
                         ,"Avg Fare": round(ride_summary["fare"].mean(),2)
                          , "Number of Rides": ride_summary["type"].count()
                         , "Type": ride_summary["type"].unique()
                    })
rides_df.head()
```

```python
urban = rides_df[rides_df["Type"]=="Urban"]
suburban = rides_df[rides_df["Type"]=="Suburban"]
rural = rides_df[rides_df["Type"]=="Rural"]
urban.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Create a legend
plt.xlabel("Number of Rides")
plt.ylabel("Average Fair")
plt.title("Ride Sharing Data")
plt.legend("Circle size correlates to driver count per city")


#print("Circle Size is an indicator of the number of drivers in the city")

# Obtain the x and y coordinates for each of the three city types
urban_x = urban["Number of Rides"]
urban_y = urban["Avg Fare"]
urban_z = urban["Number of Rides"]


suburban_x = suburban["Number of Rides"]
suburban_y = suburban["Avg Fare"]
suburban_z = suburban["Number of Rides"]

rural_x = rural["Number of Rides"]
rural_y = rural["Avg Fare"]
rural_z = rural["Number of Rides"]

# Build the scatter plots for each city types
plt.scatter(urban_x, urban_y, s=urban_z*10,  alpha=0.5, linewidth = 2, c = "blue", label = "Urban")
plt.scatter(suburban_x, suburban_y, s=suburban_z*10,  alpha=0.5, linewidth = 2, c = "red", label = "Suburban")
plt.scatter(rural_x, rural_y, s=rural_z*10,  alpha=0.5, linewidth = 2, c = "gold", label = "Rural")
plt.text(42, 40, "Note: Circle size correlates with driver count per city")
plt.legend()
#plt.show()


# Incorporate a text label regarding circle size

# Save Figure
plt.savefig("Ride Sharing.png")
```

## Total Rides by City Type

```python
#group by type
ride_summary_percents = total_ride_data.groupby(["type"])
ride_summary_percents.head(1)
```

```python
#total rides count for % calculation
total_rides= ride_data["ride_id"].count()
# total_rides
```

```python
#splitting rides by type
urban = rides_df[rides_df["Type"]=="Urban"]
suburban = rides_df[rides_df["Type"]=="Suburban"]
rural = rides_df[rides_df["Type"]=="Rural"]

# urban.head()
# suburban.head()
# rural.head()
```

```python
#calculating percentages of rides by type
urban_percent = round((urban.sum()["Number of Rides"]/total_rides)*100,2)
suburban_percent = round((suburban.sum()["Number of Rides"]/total_rides)*100,2)
rural_percent = round((rural.sum()["Number of Rides"]/total_rides)*100,2)

urban_percent
suburban_percent
rural_percent


```

```python
#labels for plotting ride percentages by type
labels = ["Urban Rides", "Suburban Rides", "Rural Rides"]
percent_of_rides = [urban_percent, suburban_percent, rural_percent]
colors = ["lightcoral", "skyblue", "gold"]
explode = (0.1, 0, 0)
```

```python
#plotting ride percentages by type
plt.title("Percent of Total Rides by City Type")
plt.pie(percent_of_rides
        , explode = explode
        , labels = labels
        , colors = colors
        , autopct = "%1.1f%%"
        , shadow =True
        , startangle = 260) 
#plt.show()
plt.savefig("Percent of Total Rides by City Type.png")
```

## Total Fares by City Type

```python
#total ride fares for % calculation
total_fares= ride_data["fare"].sum()
total_fares
```

```python
fares_summary = total_ride_data.groupby("type")
fares_summary.head()

fares_summary=round((fares_summary["fare"].sum()/total_fares)*100,2)
fares_summary
```

```python
#labels for plotting fare percentages by type
labels = ["Rural Fares", "Suburban Fares", "Urban Fares"]
percent_of_fares = fares_summary
colors = ["gold", "skyblue", "lightcoral"]
explode = (0, 0, 0.1)
percent_of_fares
```

```python
# Show Figure
#plotting ride percentages by type
plt.title("Percent of Total Fares by City Type")
plt.pie(percent_of_fares
        , explode = explode
        , labels = labels
        , colors = colors
        , autopct = "%1.1f%%"
       , shadow =True
       , startangle = 150
       ) 

#plt.show()
plt.savefig("Percent of Total Fares by City Type.png")
```

## Total Drivers by City Type

```python
#total drivers for % calculation
total_drivers= city_data["driver_count"].sum()
total_drivers
```

```python
drivers_summary = city_data.groupby("type")
drivers_summary.head()

drivers_summary=round((drivers_summary["driver_count"].sum()/total_drivers)*100,2)
drivers_summary
```

```python
# Build Pie Charts
#labels for plotting fare percentages by type
#labels = [type]
labels = ["Rural", "Suburban","Urban"]
percent_of_drivers = drivers_summary
colors = ["gold", "skyblue", "lightcoral"]
explode = (0, 0, 0.1)
percent_of_drivers
# Save Figure
```

```python
# Show Figure

#plotting driver count percentages by type
plt.title("Percent of Total Drivers by City Type")
plt.pie(percent_of_drivers
        , explode = explode
        , labels = labels
        , colors = colors
        , autopct = "%1.1f%%"
        , shadow =True
        , startangle = 180
       ) 
# plt.show()
plt.savefig("Percent of Total Drivers by City Type.png")
```
